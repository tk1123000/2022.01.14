### Pandas 활용하여 날짜 생성

```python
date = pd.date_range(start = "2021-01-01", end = "2021-03-31")
```

```python
# 문자형 포맷팅

days = date.strftime('%Y%m%d')
```

```python
days
```

```python
# 반복문의 진행상황을 체크하기 위한 라이브러리 tqdm
```

```python
from tqdm import tqdm_notebook
```

```python
# url의 바뀌는 부분을 반복작업을 통해 여러 페이지를 요청해서 수집 가능

title_list = []
score_list = []
rank_list = []
day_list = []

for day in tqdm_notebook(days):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=' + day 

    res = req.get(url)

    soup = bs(res.text,'lxml')
    title = soup.select('div.tit5 > a')
    score = soup.select('td.point')

    for i in range(len(title)):
        title_list.append(title[i].text)
        score_list.append(score[i].text)
        rank_list.append(i+1)
        day_list.append(day)
```

```python
dic = { '영화제목' : title_list, '평점' : score_list, '순위' : rank_list, '날짜' : day_list}

df = pd.DataFrame(dic)
df.set_index('순위', inplace = True)
```
